# Прогноз оттока клиентов фитнес-центра



# Описание проекта


Сеть фитнес-центров разрабатывает стратегию взаимодействия с пользователями на основе аналитических данных.

Наша задача — провести анализ анкет пользователей и подготовить план действий по удержанию клиентов.

А именно:

 - научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
 

 - сформировать типичные портреты пользователей: 
    - выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;

    - проанализировать основные признаки, наиболее сильно влияющие на отток;
    

 - сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

   - выделить целевые группы клиентов;

   - предложить меры по снижению оттока;

   - определить другие особенности взаимодействия с клиентами.

# Описание данных

Заказчик подготовил данные, которые содержат данные на месяц до оттока и факт оттока на определённый месяц. 

Набор данных включает следующие поля:
 - 'Churn' — факт оттока в текущем месяце;

Текущие поля в датасете:

1. Данные клиента за предыдущий до проверки факта оттока месяц:
 - 'gender' — пол;
 - 'Near_Location' — проживание или работа в районе, где находится фитнес-центр;
 - 'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
 - Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
 - 'Phone' — наличие контактного телефона;
 - 'Age' — возраст;
 - 'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).

2. Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
 - 'Contract_period' — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год);
 - 'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);
 - 'Group_visits' — факт посещения групповых занятий;
 - 'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;
 - 'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;
 - 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон.

#### 2.3. Построим столбчатые гистограммы и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток)<a id="step2.3"></a>

```python
#разделим посетителей на группы с признаком оттока и без него.
#напишем функцию, которая будет рисовать гистограмму распределения
left = gym.loc[gym['Churn']==1]
stayed = gym.loc[gym['Churn']==0]
def draw_distplot(column):
    sns.distplot(left[column])
    sns.distplot(stayed[column])
    plt.legend(['Отток', 'Не отток'])
    plt.show();
```

```python
# график Avg_class_frequency_total
plt.title('Частота посещений в неделю за все время действия абонемента',fontsize=12)
draw_distplot('Avg_class_frequency_total')
```

У тех, кто ушел в отток было не больше 3 посещений в неделю, в основном, 1-2. У тех кто остался в целом от 1 до 4 посещений. Те, кто остаются, ходят на тренировки чаще.

```python
# график 'Avg_class_frequency_current_month'
plt.title('Частота посещений в неделю за прошлый месяц',fontsize=12)
draw_distplot('Avg_class_frequency_current_month')
```

Большая доля тех, кто ушел в отток, в прошлом месяце вообще не посещал тренировки, то есть пришел только в этом месяце. А в остальном, также как в предыдщуем графике, те кто остаются, ходят чаще. 

```python
# график 'Avg_additional_charges_total'
plt.title('Суммарная выруска от допуслуг',fontsize=12)
draw_distplot('Avg_additional_charges_total')
```

Те, кто остаются, тратят больше, чем те, кто ушли в отток.

Построим гистограммы для столбцов с бинарными данными.

```python
countplot_col = ['gender',
 'Near_Location',
 'Partner',
 'Promo_friends',
 'Phone',
 'Group_visits']

for column in countplot_col:
    sns.countplot(x=column, hue='Churn', data=gym)
    plt.legend(['Не отток', 'отток'])
    plt.title(column)
    plt.xlabel('Значение признака')
    plt.ylabel('Количество посетителей')
    plt.grid()
    plt.show()
```

Женщин и мужчин из тех, кто ушли в отток, примерно одинаковое количество, так что пол не влияет на вероятность оттока. Среди тек, кто ушли, больше людей имеют близкое расположение к фитнес-центру, но это нельзя считать признаком вероятности оттока, так как и среди тех, кто остался - большинстов живет и работает рядом. Большинство из ушедших не являлись корпоративынми клиентами. Также, большинство из тех, кто ушли, не пришли в зал по промо. Подавляющее большинстов людей, оставляли в анкете своий номер телефона - это никак не влияет на вероятность оттока. Больше всего ушли те, кто не посещал групповые занятия. 

Посмотрим на распределение возраста, длительность посещения и длительность абонемента.

```python
countplot_col1 = ['Age',
 'Month_to_end_contract',
 'Lifetime','Contract_period']

for column in countplot_col1:
    sns.countplot(x=column, hue='Churn', data=gym)
    plt.legend(['Не отток', 'отток'])
    plt.title(column)
    plt.xlabel('Значение признака')
    plt.ylabel('Количество посетителей')
    plt.grid()
    plt.show()
```

С целевой переменной Churn сильнее всего коррелирует признак lifetime (время с момента первого обращения в фитнес-центр (в месяцах), коээфиицент корреляции Пирсона отрицательный и равен -0,44. Чем дольше человек посещает тренировки, тем меньше вероятность оттока. 
Также Churn коррелирует c количеством месяцев до окончания абонемента, размером абонемента (а последние сильно коррелируют между собой) и с частотой посещений в прошлом месяце, тк этот показатель также тесно связан с lifetime и размером абонемента. 

Есть мультиколлинеарные признаки: выше всего корреляция между количеством месяцев до окончания абонемента (Month_to_end_contract) и длительностью абонемента(Contract_period): коэффициент корреляции Пирсона равен 0.97, и  частота посещений в неделю за все время и в предыдущем месяце, Avg_class_frequency_current_month и Avg_class_frequency_total: коэффициент корреляции равен 0.95.

Есть признаки с почти нулевой корреляцией с целевой переменной: phone и gender.

 - среди тех, кто остался, больше посетителей, пришедших по промокоду друга;
 - большинство посетителей оставило свой номер в анкете;
 - в группе тех, кто остался, больше тех, кто ходит уже несколько месяец, их продолжительность тренировок выше, а среди тех, кто ушел - много новичков. Это мне кажется, важным показателем - труднее всего удержать тех, кто только начинает заниматься.
 - доля клиентов, посещавших групповые занятия, в два раза выше среди тех, кто остался;
 - возраст посетителей от 18 лет до 41 года, средний возрат по всему датасету - 29 лет. Средний возраст ушедших клиентов ниже, чем у оставшихся, но в целом, совпадает с возрастным диапазоном всех посетителей;
 - дополнительные расходы на услуги выше среди тех, кто остался, но это объясняется как раз тем, что эта выручка суммарная и оставшиеся, в среднем, ходили дольше, чем те, что ушли;
 - среди тех, кто перестал посещать фитнес-центр, контракт закончился бы через 1-2 месяца. Тем, кто продолжал посещать спортзал, оставалось в среднем около 5 месяцев до конца контракта, в среднем клиенты уходят в течение первого месяца похода в спортзал;
 - те, кто остался, посещают больше тренировок в течении недели, чем те, кто ушел в отток;
 - С целевой переменной Churn сильнее всего коррелирует признак lifetime (время с момента первого обращения в фитнес-центр (в месяцах), коээфиицент корреляции Пирсона отрицательный и равен -0,44. Чем дольше человек посещает тренировки, тем меньше вероятность оттока. Корреляция срока до окончания абонемента и длительности текущего абонемента - понятна и логична, как и корреляция средней частоты посещения в неделю за предыдущий месяц и средней частоты посещения в неделю за все время.

### 3. Шаг 3. Построим модель прогнозирования оттока клиентов<a id="step3"></a>

[К оглавлению](#main)

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце.


#### 3.1. Разобьем данные на обучающую и валидационную выборку функцией train_test_split()<a id="step3.1"></a>

Как мы увидели на матрице корреляции у нас есть две пары признаков, которые сильно коррелируют друг с другом. Чтобы избежать переученности модели нужно избавиться от мультиколлинеарности.
Из пары Contract_period - Month_to_end_contract удалим Month_to_end_contract, а из пары Avg_class_frequency_total - Avg_class_frequency_current_month удалим Avg_class_frequency_current_month.

```python
# убираем один из пары коррелирующих признаков 
gym.drop('Month_to_end_contract', axis = 1, inplace = True) 
gym.drop('Avg_class_frequency_current_month', axis = 1, inplace = True) 
```

```python
#разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = gym.drop('Churn', axis=1)
y = gym['Churn']

#разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()

#обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(X_train) 

#применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

#### 3.2. Обучим модель на train-выборке двумя способами: логистической регрессией и случайным лесом<a id="step3.2"></a>

#### Логистическая регрессия

```python
# задаем алгоритм для нашей модели
logistic_model = LogisticRegression(random_state=0, solver='lbfgs')

# обучим модель
logistic_model.fit(X_train_st, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
logistic_predictions = logistic_model.predict(X_test_st)
logistic_probabilities = logistic_model.predict_proba(X_test_st)[:,1]
```

#### Случайный лес

```python
# задаем алгоритм для нашей модели
forest_model = RandomForestClassifier(random_state = 0, n_estimators = 1000)

# обучим модель
forest_model.fit(X_train_st, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
forest_predictions = forest_model.predict(X_test_st)
forest_probabilities = forest_model.predict_proba(X_test_st)[:,1]
```

#### 3.3. Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравним по ним модели. Какая модель показала себя лучше на основании метрик?<a id="step3.3"></a>

```python
print('logistic_model:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, logistic_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, logistic_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, logistic_predictions)))
print()
print('forest_model:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, forest_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, forest_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, forest_predictions)))
```

Доля угаданных ответов одинакова у обеих моделей, полнота выше у логистической регрессии, точность выше у случайного леса. 
Так как precision и recall направлены на избежание противоположных рисков, нужна сводная метрика, учитывающая баланс между метриками. Такой метрикой является F1-score:

```python
print('f1_score для логистической регрессии: {:.3f}'.format(f1_score(y_test, logistic_predictions)))
print('f1_score для для случайного леса: {:.3f}'.format(f1_score(y_test, forest_predictions)))
```

На основании совокупности метрик лучше всего себя показала модель логистической регрессии. 

### 4.Шаг 4. Кластеризация клиентов<a id="step4"></a>

 Проведем кластеризацию объектов (клиентов):


### 4.1. Стандартизируе данные<a id="step4.1"></a>

```python
#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()

#обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(X_train) 

#применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

#### 4.2. Построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму. На основании полученного графика предположим, какое количество кластеров можно выделить<a id="step4.2"></a>

```python
# создаём объект класса scaler (нормализатор)
sc = StandardScaler()
# обучаем нормализатор и преобразуем набор данных 
x_sc = sc.fit_transform(X)
#построим матрицу расстояний
linked = linkage(x_sc, method = 'ward')
```

```python
#нарисуем дендрограмму

plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Кластеризация')
plt.show()
```

Предложительное оптимальное число кластеров 4. Однако, в условиях задачи мы договорились, что число кластеров примем за 5, то мы должны принять n = 5.

#### 4.3. Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. За число кластеров примем n_clusters=5<a id="step4.3"></a>

```python
# зададим модель k_means с количеством кластеров 5
km = KMeans(n_clusters = 5,  random_state = 0)

# спрогнозируем кластеры для наблюдений
labels = km.fit_predict(x_sc)

# сохраняем метки кластера в поле нашего датасета
gym['cluster_km'] = labels
```
